```python
import cv2
import os
```

```python
videos=sorted(os.listdir('videos'))[1:]
vid_names = [video.split('.')[0] for video in videos]
for vid_name in vid_names:
    os.makedirs(f"images/{vid_name}", exist_ok=True)
```

```python
def video_to_frames(vid_name):
    # Open the video
    video = cv2.VideoCapture(os.path.join("videos",vid_name +".mp4"))

    # Get the total number of frames
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

    # Iterate through each frame
    for i in range(total_frames):
        # Read the frame
        ret, frame = video.read()
        if not ret:
            break
        # Save the frame as an image
        cv2.imwrite(os.path.join("images",vid_name, vid_name + "_" + str(i).zfill(12) + ".jpg"), frame)

    # Release the video
    video.release()
```

```python
for vid_name in vid_names:
    video_to_frames(vid_name)
```

```python
import csv
import json
from pandas.io.json import json_normalize
import pandas as pd
```

```python
aihub_2d_pose_keypoints=pd.DataFrame()

def load_and_concat(file):
    with open(file) as data_file:
        data=json.load(data_file)
        df = json_normalize(data["people"])
        series_2d_pose_keypoints = df["pose_keypoints_2d"]
        df_2d_pose_keypoints = series_2d_pose_keypoints.to_frame()
        
        df_2d_pose_keypoints.rename(index={0:file.replace('annotations','images').replace('_keypoints.json','.jpg')},inplace=True)
        # print(df_2d_pose_keypoints)
        return_df = pd.concat([aihub_2d_pose_keypoints,df_2d_pose_keypoints])
        return return_df
```

```python
annotation_folders = sorted(os.listdir('annotations'))[1:]
for annotation_folder in annotation_folders:
    folder_dir = os.path.join('annotations',annotation_folder)
    json_files = sorted(os.listdir(folder_dir))

    for json_file in json_files:
        file_dir = os.path.join(folder_dir,json_file)
        print(file_dir)
        aihub_2d_pose_keypoints = load_and_concat(file_dir)
        
```
